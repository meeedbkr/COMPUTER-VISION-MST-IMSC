```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('..//..//assets//sudoku.png')

# Apply binary thresholding on the image
_, threshold = cv2.threshold(img, 155, 255, cv2.THRESH_BINARY)

# Convert the image to grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply adaptive thresholding using mean and Gaussian methods
mean_c = cv2.adaptiveThreshold(img_gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 15, 12)
gaus = cv2.adaptiveThreshold(img_gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 91, 12)

# Store the output images in a list
output = [img, threshold, mean_c, gaus]

# Set the titles for each output image
titles = ['Original', 'Binary', 'Adaptive Mean', 'Adaptive Gaussian']

# Display the images in a grid
for i in range(4):
    plt.subplot(2, 2, i+1)
    plt.imshow(output[i],'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
        
plt.show()
```
